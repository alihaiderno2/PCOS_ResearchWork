Categorical Symptoms Analysis

```python
categorical_cols = [
    'Weight gain(Y/N)', 'hair growth(Y/N)', 'Skin darkening (Y/N)', 
    'Hair loss(Y/N)', 'Pimples(Y/N)', 'Fast food (Y/N)', 'Reg.Exercise(Y/N)'
]

# Initailizing an empty list to hold our results
table_1_data = []

for col in categorical_cols:
    # Counting how many valid we have
    total_valid = df[col].notna().sum()
    
    yes_count = (df[col] == 1).sum()
    
    # Calculate the percentage
    yes_percent = (yes_count / total_valid) * 100
    
    formatted_string = f"{yes_count} ({yes_percent:.1f}%)"
    
    # Appending the results to our list
    table_1_data.append({
        'Clinical Symptom': col,
        'Present n (%)': formatted_string
    })
table_1_df = pd.DataFrame(table_1_data)

print("Table 1: Clinical Symptoms Summary")
display(table_1_df)
```
